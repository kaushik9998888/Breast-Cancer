```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import KaplanMeierFitter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
from sklearn.impute import SimpleImputer
```

```python
# Loading the dataset
data = pd.read_csv("Breast Cancer.csv")
```

```python
# Displaying the first few rows of the dataset
data.head()
```

```python
data.shape
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# Filling missing values in 'Cellularity' and 'Tumor Size' columns with median
data['Cellularity'].fillna(data['Cellularity'].mode()[0], inplace=True)
data['Tumor Size'].fillna(data['Tumor Size'].median(), inplace=True)
```

```python
# Dropping columns with too many missing values or irrelevant data for analysis
data_cleaned = data.drop(columns=['Patient ID', 'Cohort', 'Pam50 + Claudin-low subtype', '3-Gene classifier subtype'])
```

```python
# Converting categorical columns to numerical format using encoding
data_cleaned = pd.get_dummies(data_cleaned, drop_first=True)
```

```python
data.head()
```

```python
import warnings
warnings.filterwarnings('ignore')  # To ignore all warnings

# Handling missing values in the dataset

# Imputer for filling missing numerical values with the median
numerical_imputer = SimpleImputer(strategy='median')

# Imputer for filling missing categorical values with the mode (most frequent value)
categorical_imputer = SimpleImputer(strategy='most_frequent')

# Separating numerical and categorical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Ensure operations are performed on the original DataFrame
# Filling missing values for numerical columns
data[numerical_cols] = pd.DataFrame(numerical_imputer.fit_transform(data[numerical_cols]), columns=numerical_cols)

# Filling missing values for categorical columns
data[categorical_cols] = pd.DataFrame(categorical_imputer.fit_transform(data[categorical_cols]), columns=categorical_cols)

# Verifying that there are no missing values left
print(data.isnull().sum())
```

```python
# Selecting relevant features for model
X = data_cleaned.drop(columns=['Patient\'s Vital Status_Living', 'Overall Survival Status_Living'])
y = data_cleaned['Patient\'s Vital Status_Living']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scaling numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train_scaled, y_train)
y_pred_log = log_model.predict(X_test_scaled)
log_accuracy = accuracy_score(y_test, y_pred_log)
log_auc = roc_auc_score(y_test, y_pred_log)
```
